```python
import numpy as np
import pandas as pd

T = 5000
price = 100.0
prices = []
orders = []

for t in range(T):
    noise = np.random.randn() * 0.05
    price += noise
    prices.append(price)

    for agent_id in range(60):
        side = np.random.choice(["buy", "sell"])
        volume = np.random.randint(1, 5)
        orders.append([t, agent_id, side, price, volume])

df = pd.DataFrame(orders, columns=["t", "agent", "side", "price", "volume"])
df.to_csv("market_trace.csv", index=False)

print("Market simulation complete")
```
