```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
S0 = 100
mu = 0.08
sigma = 0.2
T = 1.0
N = 252
dt = T / N
paths = 10
```

```python
np.random.seed(42)
```

```python
S = np.zeros((paths, N + 1))
S[:, 0] = S0
```

```python
for t in range(1, N + 1):
    Z = np.random.normal(0, 1, paths)
    S[:, t] = S[:, t-1] * np.exp(
        (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
    )
```

```python
# Plot paths
plt.figure()
for i in range(paths):
    plt.plot(S[i])
plt.title("GBM Price Paths")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```

```python
# Histogram of final prices
plt.figure()
plt.hist(S[:, -1], bins=10)
plt.title("Distribution of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()
```
