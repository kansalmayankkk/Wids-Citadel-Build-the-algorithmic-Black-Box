```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/Downloads/INDIA VIX_minute.csv")
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
```

```python
df["log_return"] = np.log(df["close"] / df["close"].shift(1))
```

```python
plt.figure()
plt.hist(df["log_return"].dropna(), bins=50)
# plt.xlim(-0.05, 0.05)
plt.xlabel("Log Return")
plt.ylabel("Frequency")
plt.title("Histogram of Log Returns")
plt.show()
```

```python
window = 20   # 20 minute rolling window

df["rolling_vol"] = df["log_return"].rolling(window).std()

plt.figure()
plt.plot(df["date"], df["rolling_vol"])
plt.xlabel("Time")
plt.ylabel("Rolling Std (Volatility)")
plt.title("Volatility Clustering")
plt.show()
```
