```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/Downloads/HistoricalData_1765443493599.csv")
```

```python
## removing $ signs from closing prices to get returns
df["Close/Last"].dtype
```

```python
# removing $ sign from closing price in order to get returns eventually
df["Close/Last"] = (
    df["Close/Last"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .astype(float)
)


# datetime from string to computer comparable thing
df["Date"] = pd.to_datetime(df["Date"])

# sort
df = df.sort_values("Date")
```

```python
# daily returns
(df["Close/Last"] - df["Close/Last"].shift(1)) / df["Close/Last"].shift(1)
 
# SMA of 20 days
df["SMA_20"] = df["Close/Last"].rolling(20).mean()
```

```python
plt.figure()
plt.plot(df["Date"], df["Close/Last"])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Closing Price Over Time")
plt.show()
```
